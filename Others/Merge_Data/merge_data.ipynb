```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("iyi_antalya.csv")
df2 = pd.read_csv("kötü_antalya.csv")   
df3 = pd.read_csv("iyi_balıkesir.csv")
df4 = pd.read_csv("kötü_balıkesir.csv")   
df5 = pd.read_csv("iyi_muğla.csv")
df6 = pd.read_csv("kötü_muğla.csv")   
df7 = pd.read_csv("istanbul.csv")
df8 = pd.read_csv("eskişehir.csv")   
df9 = pd.read_csv("iyi_trabzon.csv")
df10 = pd.read_csv("kötü_trabzon.csv")   
```
