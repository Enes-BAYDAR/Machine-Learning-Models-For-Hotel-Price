```python
df = pd.read_csv('data_hotels.csv')
df.head(5)
```

```python
df.info()
```

```python
df = df.replace("Unknown", np.nan) 
df.head(5)
```

```python
nan_count = df.isna().sum()        #Accordingly, I decided to delete the entire Distance_to_Center column and other rows with NaN values.
print(nan_count)
```

```python
df = df.drop('Distance_to_Center', axis=1)
```

```python
df = df.drop('Hotel_Name', axis=1) # We won't need hotel names
```

```python
df = df.dropna()
df
```

```python
df = df.reset_index(drop=True)
```

```python
nan_count = df.isna().sum()      #Let's look at NaN values again
print(nan_count)
```

```python
df.info()
```

```python
df['Number_of_Reviews']= pd.to_numeric(df['Number_of_Reviews'])
df['Customer_Rating']= pd.to_numeric(df['Customer_Rating'])
```

```python
df['Free_Cancellation'] = df['Free_Cancellation'].replace({'Yes': 1, 'No': 0})
```

```python
df.info()
```

```python
df['Price'] = df['Price'].str.replace(" TL","")
df['Price']= pd.to_numeric(df['Price'])
df['Price']
```

```python
df.info()
```

```python
df['Accommodation_Type'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print(df['Accommodation_Type'].unique())

sns.countplot(x='Accommodation_Type',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
score_map = {
    'Ultra All Inclusive': 95,
    'All Inclusive': 85,
    'All Inclusive (Alcohol Not Included)': 75,
    'Full board': 70,
    'Half board': 60,
    'Breakfast Included': 50,
    'Standard': 40
}
df['Accommodation_Score'] = df['Accommodation_Type'].map(score_map)
```

```python
df
```

```python
df = df.drop('Accommodation_Type', axis=1)
df
```

```python
df['Hotel_Adress'].unique()
```

```python
# 'Hotel_Adress' kolonundaki verileri düzenleyerek şehir bilgisini çekelim
df['City'] = df['Hotel_Adress'].apply(lambda x: x.split(',')[-1].split('-')[0].strip())
df
```

```python
df['City'].unique()
```

```python
df['City'] = df['City'].str.replace('Trabzon Şehir Merkezi','Trabzon')
```

```python
df['City'].unique()
```

```python
df = df.drop('Hotel_Adress', axis=1) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print(df['City'].unique())

sns.countplot(x='City',data=df)
plt.xticks(rotation=90)
plt.show()
```
